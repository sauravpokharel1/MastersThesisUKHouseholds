```python

scipy.stats.levene(dftl['percentage_of_income'],dftlm['percentage_of_income'],dftum['percentage_of_income'],dftu['percentage_of_income'], center='median', proportiontocut=0.05)
```

## Paired Sample T-Test

### Transportation

```python
dft = df.loc[df['Category'].isin(['transportation'])]
dftl = dft.loc[dft['income_group'].isin(['Lower Class','Lower Middle Class'])] #lowerclass
dftl1 = dftl.loc[dftl['year'] == 2019]
dftl2 = dftl.loc[dftl['year'] == 2020]
dftu = dft.loc[dft['income_group'].isin(['Upper Class','Upper Middle Class'])]
dftu1 = dftu.loc[dftu['year'] == 2019]
dftu2 = dftu.loc[dftu['year'] == 2020]
```

```python
stats.ttest_rel(dftl1['percentage_of_income'], dftl2['percentage_of_income'])
```

```python
scipy.stats.wilcoxon(dftl1['percentage_of_income'], dftl2['percentage_of_income'])
```

```python
stats.ttest_rel(dftu1['percentage_of_income'], dftu2['percentage_of_income'])
```

```python
scipy.stats.wilcoxon(dftu1['percentage_of_income'], dftu2['percentage_of_income'])
```

Paired t-test
1. H4 : null hypothesis of equal averages cannot be rejected for lower income households' spent in transportation before and after pandemic. 
2. H8 : null hypothesis of equal averages cannot be rejected for higher income households' spent in transportation before and after pandemic

Wilcoxon Signed-rank Test
1. H4 : null hypothesis of equal averages cannot be rejected for lower income households' spent in transportation before and after pandemic. 
2. H8 : null hypothesis of equal averages cannot be rejected for higher income households' spent in transportation before and after pandemic

Both the tests show that there was no statistically significant effect of pandemic on income spent on transportation to both upper and lower income households.
